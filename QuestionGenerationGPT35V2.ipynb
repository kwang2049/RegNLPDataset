```python
import json
import os
import openai
import time
import re
import random
import string

# Set environment variables for API keys
openai.api_key = "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"

def generate_question(context):
        questions = []
    #while True:  # Keep trying until successful
        try:
            response = openai.ChatCompletion.create(
                model="gpt-3.5-turbo",
                messages=[
                    {"role": "system", "content": "Your task is to generate realistic and applied questions that pertains to the provided regulatory or compliance material. Ensure that the context implicitly contains the answer to the question."},
                    {"role": "user", "content": context}
                ]
            )
            content = response.choices[0].message["content"].strip()
            # Check if the response is a question; if not, return 'None'
            questions = [q.strip() for q in content.split('\n') if q.endswith('?')]
            questions = [re.sub(r'^\d+\.\s+', '', q) for q in questions]
        except Exception as e:
            print(f"An error occurred: {e}")
            print("Retrying in 2 minutes...")
            time.sleep(120)  # Wait for 120 seconds before retrying
        print(f"Question count : {len(questions)}")
        return questions

def process_data_and_save(data, file_path):
    document_id = "1"
    generated_ids = set()  # A set to keep track of generated IDs

    for item in data:  # Start processing from the 401st item
        obligations = item.get('Obligations', [])
        passage = item.get('Text', '').strip()
        hierarchical_info = item.get('ContextID', '').strip()

        if obligations:  # Check if 'Obligations' is not empty
            questions = generate_question(passage)  # Assuming this function returns a list of questions
            for question in questions:
                question_id = generate_unique_id(generated_ids)
                passage_id = generate_unique_id(generated_ids)

                qa_pair = {
                    "DocumentID": document_id,
                    "QuestionID": question_id,
                    "Question": question,
                    "PassageID": passage_id,
                    "Passage": passage,
                    "HierarchicalInfo": hierarchical_info
                }

                save_to_json(qa_pair, file_path)

def generate_unique_id(existing_ids):
    while True:
        new_id = ''.join(random.choices(string.digits, k=3)) + ''.join(random.choices(string.ascii_letters, k=3))
        if new_id not in existing_ids:
            existing_ids.add(new_id)
            return new_id

def save_to_json(qa_pair, file_path):
    try:
        with open(file_path, 'r+') as file:
            # Read the existing data
            file_data = json.load(file)
            # Append the new QA pair
            file_data.append(qa_pair)
            # Reset file pointer to the beginning
            file.seek(0)
            # Write the updated data back to the file
            json.dump(file_data, file, indent=4)
            # Print the count of items in the JSON file
            print(f"Item count in the JSON file after save: {len(file_data)}")
    except FileNotFoundError:
        with open(file_path, 'w') as file:
            # If the file does not exist, create a new list with the QA pair and write it to the file
            json.dump([qa_pair], file, indent=4)
            # Since this is the first item, print count as 1
            print("Item count in the JSON file after save: 1")

def main():
    # Load data from JSON
    file_path = '/content/CoBs_Obligations_Named_Entities_Defined_Terms (1).json'
    with open(file_path, 'r') as file:
        data = json.load(file)

    output_file_path = '/content/drive/Othercomputers/MBZUAI/MBZUAI/ADGM-Project/ADGM Docs Study/ArtificialDataset/COBS_VER15.150823_raw_questions.json'
    process_data_and_save(data, output_file_path)

    print(f"Questions and contexts saved to {output_file_path}")

if __name__ == "__main__":
    main()
```
