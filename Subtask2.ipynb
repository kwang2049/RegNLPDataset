```python
import json
import openai
import os

class QnASystem:
    # Static JSON file path
    SOURCE_JSON_FILE_PATH = "/content/drive/Othercomputers/MBZUAI/MBZUAI/ADGM-Project/SharedTask/CoBs_Raw_Data.json"

    def __init__(self):
        # Load the JSON data once during initialization
        self.data = self.load_json_data(QnASystem.SOURCE_JSON_FILE_PATH)

    @staticmethod
    def load_json_data(file_path):
        with open(file_path, 'r', encoding='utf-8') as file:
            return json.load(file)

    def get_relevant_context(self, question, max_tokens=1000):
        context = ""
        keywords = question.split()
        question_tokens = len(question.split())
        instruction_tokens = 20
        available_tokens = max_tokens - question_tokens - instruction_tokens

        for item in self.data:
            passage_tokens = len(item['Passage'].split())
            if any(keyword.lower() in item['Passage'].lower() for keyword in keywords) and passage_tokens < available_tokens:
                context += item['Passage'] + " "
                available_tokens -= passage_tokens

        return context

    def ask_gpt_with_context(self, question, context, model="gpt-3.5-turbo"):
        openai.api_key = "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"  # Use environment variable for API key
        messages = [{"role": "system", "content": context},
                    {"role": "user", "content": question}]

        response = openai.ChatCompletion.create(
            model=model,
            messages=messages
        )

        return response.choices[0].message['content'].strip()

    def generate_answer(self, question, max_tokens=6000):
        context = self.get_relevant_context(question, max_tokens)
        print("\nGenerating answer with context...")
        answer = self.ask_gpt_with_context(question, context)
        print(answer)

def load_json_data(file_path):
        with open(file_path, 'r', encoding='utf-8') as file:
            return json.load(file)

# Function to process each question and generate the answer
def process_questions(questions):
    qna_system = QnASystem()
    results = []
    for index, item in enumerate(questions, start=1):  # Use enumerate for item count
        question_id = item.get("QuestionID")
        question = item.get("Question")

        # Generate answer using GPT-3.5-turbo
        answer = qna_system.generate_answer(question)

        # Append the result
        results.append({
            "QuestionID": question_id,
            "Question": question,
            "Answer": answer
        })

        # Print the item count
        print(f"Processed {index}/{len(questions)} questions.")

    return results

# Function to save the results in a new JSON file
def save_results(results, file_path):
    with open(file_path, 'w', encoding='utf-8') as file:
        json.dump(results, file, ensure_ascii=False, indent=4)



if __name__ == "__main__":
    questions_file_path = "/content/drive/Othercomputers/MBZUAI/MBZUAI/ADGM-Project/SharedTask/COBS_VER15.150823_entailed_questions.json"  # Change this to your file path
    # Load the questions from the JSON file
    questions_data = load_json_data(questions_file_path)

    # Process each question and generate answers
    processed_data = process_questions(questions_data)

    new_json_file_path="/content/drive/Othercomputers/MBZUAI/MBZUAI/ADGM-Project/SharedTask/COBS_VER15.150823_answer_entailed_questions_subtask2.json"
    # Save the processed data in a new JSON file
    save_results(processed_data, new_json_file_path)

    print(f"Processed data has been saved to {new_json_file_path}")

```
